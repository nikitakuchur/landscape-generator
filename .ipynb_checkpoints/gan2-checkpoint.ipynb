```python
import torch
from torch import nn
import torch.nn.functional as F
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader

from PIL import Image
import os
```

```python
class ImagesDataset(Dataset):
    
    def __init__(self, root_dir):
        self.root_dir = root_dir
        self.img_names = os.listdir(root_dir)
        self.data_transforms = transforms.Compose([
            transforms.RandomHorizontalFlip(),
            transforms.RandomVerticalFlip(),
            transforms.ToTensor()])
        
    def __getitem__(self, index):
        img = Image.open(self.root_dir + '/' + self.img_names[index]).convert('L')
        item = self.data_transforms(img)
        img.close()
        min = item.min()
        max = item.max()
        if min != max:
            item = (item - min) / (max - min)
        return item.view(1, 64, 64)

    def __len__(self):
        return len(self.img_names)
```

```python
import matplotlib.pyplot as plt

dataset = ImagesDataset('./data')
p = plt.imshow(dataset[3600].view(64, 64))
plt.colorbar(p)
plt.show()
```

```python
batch_size = 100
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
```

```python
def cnn(in_n, out_n):
    return nn.Conv2d(in_n, out_n, kernel_size=3, padding=1)
```

```python
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.cnn_1 = cnn(1, 32)
        self.cnn_2 = cnn(32, 64)
        self.cnn_3 = cnn(64, 128)
        self.cnn_4 = cnn(128, 256)
        self.cnn_5 = cnn(256, 1)
        
    def forward(self, data):
        x = self.cnn_1(data)
        x = F.leaky_relu(F.max_pool2d(x, 2))
        
        x = self.cnn_2(x)
        x = F.leaky_relu(F.max_pool2d(x, 2))
        
        x = self.cnn_3(x)
        x = F.leaky_relu(F.max_pool2d(x, 2))
        
        x = self.cnn_4(x)
        x = F.leaky_relu(F.max_pool2d(x, 8))
        
        x = torch.sigmoid(self.cnn_5(x))
        return x
```

```python
disc = Discriminator().cuda()
```

```python
x = dataset[400].cuda()
out = disc(x.view(1, 1, 64, 64))
print(out.item())
```

```python
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.cnn_1 = cnn(2, 32)
        self.cnn_2 = cnn(32, 64)
        self.cnn_3 = cnn(64, 128)
        self.cnn_4 = cnn(128, 256)
        self.cnn_5 = cnn(256, 128)
        self.сnn_6 = cnn(128, 64)
        self.cnn_7 = cnn(64, 32)
        self.cnn_8 = cnn(32, 1)
        
    def forward(self, data, mask):
        x = torch.cat((data, mask), 1)
        
        x = self.cnn_1(x)
        x = F.leaky_relu(F.max_pool2d(x, 2))
        x = self.cnn_2(x)
        x = F.leaky_relu(F.max_pool2d(x, 2))
        x = self.cnn_3(x)
        x = F.leaky_relu(F.max_pool2d(x, 2))
        
        x = self.cnn_4(x)
        x = F.leaky_relu(F.max_pool2d(x, 2))
                         
        x = F.interpolate(x, 8)
        x = F.leaky_relu(self.cnn_5(x))
        
        x = F.interpolate(x, 16)
        x = F.leaky_relu(self.сnn_6(x))
        x = F.interpolate(x, 32)
        x = F.leaky_relu(self.cnn_7(x))
        x = F.interpolate(x, 64)
        x = torch.sigmoid(self.cnn_8(x))
        return x
```

```python
gen = Generator().cuda()
```

```python
def gen_broken(size):
    broken = torch.zeros(size, 4, dtype=torch.bool).cuda()
    for i in range(size):
        randInt = torch.IntTensor(1).random_(0, 4).cuda().item()
        broken[i][randInt] = True
        randFloat = torch.randn(1).cuda().item()
        if randFloat > 0:
            broken[i][(randInt + 1) % 4] = True
    return broken

def damage_part(tensor, mask, broken, x, y):
    if broken == True:
        tensor[0, x:x + 32, y:y + 32] = torch.rand(32, 32).cuda()
        mask[0, x:x + 32, y:y + 32] = 1

def damage(tensor, broken):
    tensor = tensor.view(-1, 1, 64, 64).clone()
    mask = torch.zeros(tensor.shape[0], 1, 64, 64).cuda()
    for i in range(tensor.shape[0]):
        damage_part(tensor[i], mask[i], broken[i][0], 0, 0)
        damage_part(tensor[i], mask[i], broken[i][1], 0, 32)
        damage_part(tensor[i], mask[i], broken[i][2], 32, 32)
        damage_part(tensor[i], mask[i], broken[i][3], 32, 0)
    return tensor, mask

def past_part(tensor, gen_tesor, broken, x, y):
    if broken == True:
        tensor[0, x:x + 32, y:y + 32] = gen_tesor[0, x:x + 32, y:y + 32]

def past(tensor, gen_tesor, broken):
    tensor = tensor.view(-1, 1, 64, 64).clone()
    gen_tesor = gen_tesor.view(-1, 1, 64, 64)
    for i in range(tensor.shape[0]):
        past_part(tensor[i], gen_tesor[i], broken[i][0], 0, 0)
        past_part(tensor[i], gen_tesor[i], broken[i][1], 0, 32)
        past_part(tensor[i], gen_tesor[i], broken[i][2], 32, 32)
        past_part(tensor[i], gen_tesor[i], broken[i][3], 32, 0)
    return tensor
```

```python
x = dataset[6870].cuda()
broken = gen_broken(1)

print(broken)
damaged_x, mask_x = damage(x, broken)

out = gen(damaged_x, mask_x)
reconst_tensor = past(damaged_x, out, broken)

plt.imshow(damaged_x.view(64, 64).cpu().detach().numpy())
plt.show()

plt.imshow(reconst_tensor.view(64, 64).cpu().detach().numpy())
plt.show()
```

```python
# Generator
def generate():
    big_tensor = torch.rand(64, 256).cuda()

    big_tensor[0:64, 0:64] = dataset[6171].cuda()

    mask = torch.zeros(64, 64).cuda()
    mask[0:64, 32:64] = 1

    i = 1
    while i < 7:
        x = i * 32
        part = big_tensor[0:64, x:x + 64].cuda()
        out = gen(part.view(1, 1, 64, 64), mask.view(1, 1, 64, 64))
        big_tensor[0:64, x + 32:x + 64] = out[0, 0, 0:64, 32:64]
        i += 1

    plt.imshow(big_tensor.view(64, 256).cpu().detach())
    plt.show()
```

```python
import torch.optim as optim

disc_optimizer = optim.Adam(disc.parameters(), lr=0.001, betas=(0.5, 0.999))
gen_optimizer = optim.Adam(gen.parameters(), lr=0.001, betas=(0.5, 0.999))
```

```python
disc.train()
gen.train()

epochs = 100

loss_func = nn.BCELoss()

for epoch in range(epochs):
    for i, data in enumerate(dataloader):
        data = data.cuda()
        data_size = data.shape[0]
        
        # Train discriminator
        
        # real data
        disc.zero_grad()
        label = (torch.ones(data_size) * 0.9).cuda()
        out = disc(data).reshape(-1)
        disc_loss_real = loss_func(out, label)
        
        # fake data
        broken = gen_broken(data_size)
        damaged_data, mask = damage(data, broken)
        gen_data = gen(damaged_data, mask)
        fake_data = past(damaged_data, gen_data, broken)
        label = (torch.ones(data_size) * 0.1).cuda()
        out = disc(fake_data.detach()).reshape(-1)
        disc_loss_fake = loss_func(out, label)
        
        disc_loss = disc_loss_real + disc_loss_fake
        disc_loss.backward()
        disc_optimizer.step()
        
        # Train generator
        gen.zero_grad()
        label = torch.ones(data_size).cuda()
        out = disc(fake_data).reshape(-1)
        gen_loss = loss_func(out, label)
        gen_loss.backward()
        gen_optimizer.step()
        
        if i % 100 == 0:
            plt.imshow(damaged_data[0][0].view(64, 64).cpu().detach().numpy())
            plt.show()
            plt.imshow(fake_data[0][0].view(64, 64).cpu().detach().numpy())
            plt.show()
            generate()
            
            msg = "epoch[{}/{}], disc_loss={:.3f}, gen_loss={:.3f}".format(epoch + 1, epochs, 
                    disc_loss.item(), gen_loss.item())
            print(msg)
```
